```python
import pickle
import numpy as np
import itertools
import pandas as pd
from sklearn.metrics import accuracy_score

darray=[]

filestr = '/mnt/amelia_data/5classes_new/10percent_negative_30iter.pkl'

f = pickle.load(open(filestr,'rb'))

for loop in range(len(f)):
    A = np.dstack((f[loop][0],f[loop][1],f[loop][2],f[loop][3],f[loop][4],f[loop][5],f[loop][6],f[loop][7],f[loop][8],f[loop][9]))
    var = np.std(A, axis=2)
    darray.append(np.ravel(var[0]))
list_flat_negative = list(itertools.chain(*darray))    

darray=[]
pred_train_label=[]

filestr = '/mnt/amelia_data/5classes_new/10percent_train_30iter.pkl'
f = pickle.load(open(filestr,'rb'))

for loop in range(len(f)):
    A = np.dstack((f[loop][0],f[loop][1],f[loop][2],f[loop][3],f[loop][4],f[loop][5],f[loop][6],f[loop][7],f[loop][8],f[loop][9]))
    var = np.std(A, axis=2)
    mean_train = np.mean(A, axis=2)
    darray.append(np.ravel(var[0]))
    pred_train_label.append(np.argmax(mean_train))
list_flat_train = list(itertools.chain(*darray))

darray=[]
pred_test_label=[]

filestr = '/mnt/amelia_data/5classes_new/10percent_test_30iter.pkl'
f = pickle.load(open(filestr,'rb'))

for loop in range(len(f)):
    A = np.dstack((f[loop][0],f[loop][1],f[loop][2],f[loop][3],f[loop][4],f[loop][5],f[loop][6],f[loop][7],f[loop][8],f[loop][9]))
    var = np.std(A, axis=2)
    mean_test = np.mean(A, axis=2)
    darray.append(np.ravel(var[0]))
    pred_test_label.append(np.argmax(mean_test))

list_flat_test = list(itertools.chain(*darray))

df_train = pd.read_csv('/mnt/amelia_data/381_5_classes/split0/train.tsv', delimiter='\t', header=None)
df_train.columns = ['Questions','numeric label']
df_train= df_train[pd.notnull(df_train['Questions'])]
df_train['Questions'] = df_train['Questions'].astype(str)
train_labels = df_train['numeric label'].values

df_test = pd.read_csv('/mnt/amelia_data/381_5_classes/split0/test.tsv', delimiter='\t', header=None)
df_test.columns = ['Questions','numeric label']
df_test['numeric label'] = df_test['numeric label'].astype(np.int64)
test_labels = df_test['numeric label'].values


print(accuracy_score(pred_train_label,train_labels))
print(accuracy_score(pred_test_label,test_labels))
```

```python
# grid search
import random
from random import randint 
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score

# try on new data 

v_10_train = pickle.load(open('/mnt/amelia_data/5classes_new/10percent_train.pkl','rb'))
v_10_test = pickle.load(open('/mnt/amelia_data/5classes_new/10percent_test.pkl','rb'))
v_10_negative = pickle.load(open('/mnt/amelia_data/5classes_new/10percent_negative.pkl','rb'))


negative_mean = np.mean(v_10_negative,1)
negative_variance = np.std(v_10_negative,1)


# test_mean, test_variance = pickle.load(open('/mnt/amelia_data/5classes/20percent_test.pkl','rb'))
# negative_mean, negative_variance = pickle.load(open('/mnt/amelia_data/5classes/20percent_negative.pkl','rb'))
# escalation_mean, escalation_variance = pickle.load(open('/mnt/amelia_data/5classes/20percent_escalation.pkl','rb'))


# obtain test labels
df_test = pd.read_csv('/mnt/amelia_data/381_5_classes/split0/test.tsv', delimiter='\t', header=None)
df_test.columns = ['Questions','numeric label']
df_test['numeric label'] = df_test['numeric label'].astype(np.int64)
tlabel = df_test['numeric label'].values


#tlabel = pickle.load(open('/mnt/amelia_data/381classes/test_labels.pkl','rb'))

test_mean = np.mean(v_10_test,1)
test_variance = np.std(v_10_test,1) 

num_test = 906

index_test = random.sample(range(len(tlabel)), num_test)
tlabel = tlabel[index_test]
test_mean = test_mean[index_test]
test_variance = test_variance[index_test,:]




negative_mean = np.mean(v_10_negative,1)
negative_variance = np.std(v_10_negative,1)



num_negative = 3500

index_negative= random.sample(range(len(negative_mean)), num_negative)
uc_negative = negative_variance[index_negative,:]
mp_negative = negative_mean[index_negative,:]



alllabels= np.concatenate((tlabel, np.ones(len(negative_mean))*5))

# A is the matrix of mean probability
# B is the matrix of uncertainty





tl_negative = np.ones(num_negative)*5

tl = np.concatenate((tlabel,  tl_negative), axis=0)
#pl = np.concatenate((pl_split0, pl_escalation, pl_negative), axis=0)
uc = np.concatenate((test_variance,  uc_negative), axis=0)
mp = np.concatenate((test_mean, mp_negative), axis=0)

A= np.zeros((len(mp),6))*0
B= np.ones((len(mp),6))*0


A[:,:-1]=mp
B[:,:-1]=uc


def varianceCut_2(mp, mv, cutp, cutv):
    x_mat = np.zeros((len(mp), 6))
    x_label=np.zeros(len(mp))
    
    for loop in range(len(mp)):
        assignFlag = False    
        
        for cloop in range(5):
            if ((mp[loop, cloop]>=cutp) and (mv[loop, cloop]<=cutv)):
                x_mat[loop,cloop]=1
                assignFlag = True
            
        if not assignFlag:
            x_mat[loop, 5]=1
                
    for loop in range(len(mp)):
        x_label[loop]=np.argmax(x_mat[loop])
        
    return x_label

xx= np.linspace(0,1,100)
yy = np.linspace(0,1,100)

def fscore(x, y):
    r = np.zeros((len(x),len(y)))
    for i in range(len(x)):
        for j in range(len(y)):
            test_prediction = varianceCut_2(A, B, x[i], y[j])
            binary_pred =  np.where(test_prediction < 5, 1, -1)
            binary_true =  np.where(tl < 5, 1, -1)
            r[i,j]=f1_score(binary_pred, binary_true)
    return r

grid = fscore(xx,yy)
```
